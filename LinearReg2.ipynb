```python
import pandas as pd 
import numpy as np 
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
data = pd.concat([train,test])
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe().columns
```

```python
cathegorical_data = data.select_dtypes(include="object")
```

```python
numeric_data = data.select_dtypes(include=["float64","int64"])
```

```python
numeric_data.head()
```

```python
not_importent = numeric_data.corr()["SalePrice"][numeric_data.corr()["SalePrice"].values < 0.3 ].keys()
```

```python
numeric_data = numeric_data.drop(not_importent,axis=1)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
itr_imp = IterativeImputer(random_state=0)

#knn_imp = KNNImputer(n_neighbors=15, weights="uniform")
```

```python
imputed_numeric = pd.DataFrame(itr_imp.fit_transform(numeric_data.drop("SalePrice",axis=1)),columns=numeric_data.drop("SalePrice",axis=1).columns)
```

```python
imputed_numeric["SalePrice"] = numeric_data["SalePrice"].values
```

```python
pd.concat([imputed_numeric.corr()["SalePrice"],numeric_data.corr()["SalePrice"]],axis=1)
```

```python
ordinal_data =  cathegorical_data[["BsmtQual","BsmtCond","BsmtExposure","ExterQual","ExterCond","HeatingQC" \
     ,"KitchenQual","FireplaceQu","GarageQual","GarageCond","PoolQC","BsmtFinType1","BsmtFinType2","Functional"]]
```

```python
ordinal_data.isna().sum()
```

```python
from sklearn.impute import SimpleImputer

s_imp = SimpleImputer(strategy="most_frequent")

imputed_ordinals = s_imp.fit_transform(ordinal_data)
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()

encoded_ordinals = oe.fit_transform(imputed_ordinals)
```

```python
encoded_ordinal_data = pd.DataFrame(encoded_ordinals,columns=ordinal_data.columns)
```

```python
encoded_ordinal_data.head()
```

```python
encoded_ordinal_data.shape
```

```python
numeric_data.shape
```

```python
left_data = cathegorical_data.drop(ordinal_data.columns,axis=1)
```

```python
imputed_left = s_imp.fit_transform(left_data)
imputed_left
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop="first")
```

```python
encoded_left = ohe.fit_transform(imputed_left)
```

```python
encoded_left_data = pd.DataFrame(encoded_left.toarray())
encoded_left_data.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 1,random_state=0)
pca_values = pca.fit_transform(encoded_left_data)
pca_data = pd.DataFrame(pca_values)
```

```python
pca_data["SalePrice"]= numeric_data["SalePrice"].values
```

```python
pca_data.corr()["SalePrice"]
```

```python
last_data = pd.concat([imputed_numeric,encoded_ordinal_data,pca_data],axis=1)
```

```python
last_data.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_last_data = scaler.fit_transform(last_data.drop("SalePrice",axis=1))
```

```python
scaled_last_data = pd.DataFrame(scaled_last_data,columns=last_data.drop("SalePrice",axis=1).columns)
scaled_last_data["SalePrice"] = numeric_data["SalePrice"].values
scaled_last_data.head()
```

```python
train_data = scaled_last_data[:1460]
test_data = scaled_last_data[1460:]
```

```python
X_train = train_data.drop(["SalePrice"],axis = 1)
y_train = train_data["SalePrice"]

X_test = test_data.drop(["SalePrice"],axis = 1)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
output = pd.DataFrame({'Id': test['Id'].values, 'SalePrice': predictions})
output.to_csv('submission2.csv',index = False )
```
